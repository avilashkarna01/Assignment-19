```python
#Q.1 - Create a numpy array with 10 elements of the shape(10,1) using np.random and
#find out the mean of the elements using basic numpy functions.
import numpy as np
a=np.random.random([10,1])
print(a)
print("mean=")
print(a.mean())
print("\n")
```

```python
# Q.2 - Create a numpy array with 20 elements of the shape(20,1) using np.random
# find the variance and standard deviation of the elements.
import numpy as np
a=np.random.random([20,1])
print(a)
print("var=")
print(a.var())
print("std=")
print(a.std())
print("\n")
```

```python
#Q.3 - Create a numpy array A of shape(10,20) and B of shape (20,25) using np.random.
# Print the matrix which is the matrix multiplication of A and B.
# The shape of the new matrix should be (10,25). Using basic numpy math functions only
# find the sum of all the elements of the new matrix.

import numpy as np
x=np.random.random([10,20])
y=np.random.random([20,25]) 
z=np.random.random([10,25])
print(x)
print(y)
print(z)
print("multiplication of matrix")
print(np.matmul(x,y))
print("sum of all the elements of new matrix z")
print(np.sum(z))
print("/n")
```
